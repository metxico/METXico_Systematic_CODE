```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('InputNameOFFile.csv')
```

```python
molecules = [Chem.MolFromSmiles(smiles) for smiles in df['ColumnNameOFsmiles']]
```

```python
descriptors = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024) for mol in molecules]
```

```python
descriptors_list = [list(desc) for desc in descriptors]
```

```python
pca = PCA(n_components=2)
reduced_descriptors = pca.fit_transform(descriptors_list)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(reduced_descriptors[:, 0], reduced_descriptors[:, 1], alpha=0.7)
plt.title("OutputNameOFplot")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
from sklearn.manifold import TSNE
```

```python
descriptors_array = np.array(descriptors_list)
```

```python
tsne = TSNE(n_components=3, perplexity=30, n_iter=300)
reduced_descriptors_3d = tsne.fit_transform(descriptors_array)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(reduced_descriptors_3d[:, 0], reduced_descriptors_3d[:, 1], reduced_descriptors_3d[:, 2], alpha=0.7)
ax.set_title("outputNameOFplot")
ax.set_xlabel("t-SNE 1")
ax.set_ylabel("t-SNE 2")
ax.set_zlabel("t-SNE 3")
plt.show()
```
