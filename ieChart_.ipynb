```python
import pandas as pd

# Leer el archivo CSV subido
df = pd.read_csv('InputOFfilename.csv')

# Revisar los primeros datos
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
grouped_data = df['ColumnNameOFinterest'].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
```

```python
plt.title("Distribución de ClassyFire Superclass")
plt.tight_layout()
```

```python
plt.show()
```
