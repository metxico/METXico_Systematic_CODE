```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_name = "fileNameOFinterest.csv"
data = pd.read_csv(file_name)
```

```python
mz_values = data.iloc[:, 0]  # 'The name of first column'
data_clean = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').fillna(0)  # Convertir a num√©ricos y eliminar NaNs
```

```python
plt.figure(figsize=(12, 8))
sns.clustermap(data_clean, cmap="viridis", metric="euclidean", method="ward",
               standard_scale=1, yticklabels=mz_values)
```

```python
plt.title("Clustering Heatmap de √Åreas de Picos con m/z en Eje Y")
plt.show()
```
